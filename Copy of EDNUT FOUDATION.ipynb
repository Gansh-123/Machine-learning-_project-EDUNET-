```python
from sklearn.preprocessing import LabelEncoder
Encoder=LabelEncoder()
df["Speciess"]=Encoder.fit_transform(df["Species"])
df.drop("Species",axis=1,inplace=True)
df
```

```python
x=df.drop("Speciess",axis=1)
print(x)
```

```python
y=df["Speciess"]
print(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max=MinMaxScaler()
Scaled_df=min_max.fit_transform(df)
new_data=pd.DataFrame(Scaled_df,columns=df.columns)
new_data
```

```python
plt.boxplot(x=new_data[["SepalLengthCm","SepalWidthCm","PetalLengthCm"]],labels=["SpealLengthCm","Spealwidth","PetalLengthCm"])
plt.legend()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
model=LogisticRegression()
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=56)
model.fit(x_train,y_train)
#predication
predict_value=model.predict(x_test)
print(predict_value)
```

```python
from sklearn.metrics import accuracy_score, precision_score,f1_score
accuracy_scoe=accuracy_score(y_test,predict_value)
print(accuracy_scoe*100)
precision_score=precision_score(y_test,predict_value, average="micro")
print(precision_score)
f1_score=f1_score(y_test,predict_value,average="micro")
print(f1_score)
```
